# Testes modelo ampacidade

```python
from normaslt import geral
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi
```

```python
# Parâmetros básicos 
cabo = geral.dbCabo.CAA_HAWK
amp = 800
vvento = 0.5
tempAr = 30.
radglobal = 800.
alt = 500.
rugcond = 0.9
absorcao = 0.6
emicond = 0.6
```

```python
# Teste 1 - temperatura ambiente
vecT = np.linspace(0, 45, 50)
t1 = np.zeros(len(vecT))
for i,t in enumerate(vecT):
    t1[i] = geral.tempCondutorCigre(amp, vvento, pi/2, t, radglobal, alt, cabo, rugcond, absorcao, emicond)
plt.plot(vecT, t1)
plt.xlabel('Temperatura ambiente (°C)')
plt.ylabel('Temperatura condutor (°C)')
plt.grid(True)
plt.show()
```

```python
# Teste 2 - velocidade do vento
vecV = np.logspace(-1, 1, 100)
t1 = np.zeros(len(vecV))
for i,v in enumerate(vecV):
    t1[i] = geral.tempCondutorCigre(amp, v, pi/2, tempAr, radglobal, alt, cabo, rugcond, absorcao, emicond)
plt.semilogx(vecV, t1)
plt.xlabel('Velocidade do vento (m/s)')
plt.ylabel('Temperatura condutor (°C)')
plt.grid(True, which='both')
plt.show()
```

```python
# Teste 3 - radiação solar
vecR = np.linspace(0, 1200, 50)
t1 = np.zeros(len(vecR))
for i,r in enumerate(vecR):
    t1[i] = geral.tempCondutorCigre(amp, vvento, pi/2, tempAr, r, alt, cabo, rugcond, absorcao, emicond)
plt.plot(vecR, t1)
plt.xlabel('Radiação solar (W/m²)')
plt.ylabel('Temperatura condutor (°C)')
plt.grid(True)
plt.show()
```

```python
# Teste 4 - altitude
vecA = np.linspace(0, 1500, 50)
t1 = np.zeros(len(vecA))
for i,a in enumerate(vecA):
    t1[i] = geral.tempCondutorCigre(amp, vvento, pi/2, tempAr, radglobal, a, cabo, rugcond, absorcao, emicond)
plt.plot(vecA, t1)
plt.xlabel('Altitude (m)')
plt.ylabel('Temperatura condutor (°C)')
plt.grid(True)
plt.show()
```
