# Projeto de geometria de torre 69 kV
Comparação entre critérios de norma

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
from math import degrees, radians, sin, cos, sqrt
from normaslt import geral
from normaslt.geral import plot_dimension
```

```python
# Dados de projeto
v = 69.
cond = geral.dbCabo.CAA_GROSBEAK
Vb = 32 # Velocidade básica do vento, m/s
```

```python
# Teste plotagem - torre 69 kV

# Posição das fases
xr = 1.5*cos(radians(30))
x1 = xr
x2 = -xr
x3 = xr
# Alturas relativas à fase mais baixa
y1 = 0.
y2 = 1.5
y3 = 3.
lc = 1. # Comprimento da cadeia (m)
f = 8. # Flecha (m)
h = 10. # clearance (m)
# Ângulos de balanço
ang1 = radians(25.)
ang2 = radians(45.)
hbase = h + f + lc
y1t = y1+hbase
y2t = y2+hbase
y3t = y3+hbase

# Estrutura
ltsup = 0.3
ltinf = 0.6
ht = y3t + 0.5

plt.figure()
plt.plot([-ltinf/2, -ltsup/2], [0, ht],'k-')
plt.plot([ltinf/2, ltsup/2], [0, ht],'k-')
plt.plot([-ltsup/2, ltsup/2], [ht, ht],'k-')
plt.plot([0, x1], [y1t, y1t],'k-')
plt.plot([0, x2], [y2t, y2t],'k-')
plt.plot([0, x3], [y3t, y3t],'k-')
plt.plot([-4.5, 4.5], [0, 0],'k-', linewidth=0.5)
plt.plot([x1, x1], [y1t, y1t-lc],'bo-')
plt.plot([x2, x2], [y2t, y2t-lc],'bo-')
plt.plot([x3, x3], [y3t, y3t-lc],'bo-')
plt.plot([x1, x1], [y1t-0.2, y1t-lc+0.2],'g-', linewidth=3)
plt.plot([x2, x2], [y2t-0.2, y2t-lc+0.2],'g-', linewidth=3)
plt.plot([x3, x3], [y3t-0.2, y3t-lc+0.2],'g-', linewidth=3)
plt.plot([x1, x1-lc*sin(ang1)], [y1t, y1t-lc*cos(ang1)],'ro-')
plt.plot([x2, x2+lc*sin(ang1)], [y2t, y2t-lc*cos(ang1)],'ro-')
plt.plot([x3, x3-lc*sin(ang1)], [y3t, y3t-lc*cos(ang1)],'ro-')
plt.gca().add_patch(patches.Arc((x1, y1t), 2*lc, 2*lc, theta1=-degrees(ang1)-90., theta2=-90.,lw=0.5,color='r'))
plt.gca().add_patch(patches.Arc((x2, y2t), 2*lc, 2*lc, theta1=-90., theta2=degrees(ang1)-90.,lw=0.5,color='r'))
plt.gca().add_patch(patches.Arc((x3, y3t), 2*lc, 2*lc, theta1=-degrees(ang1)-90., theta2=-90.,lw=0.5,color='r'))
# Espaçamentos
dist1 = 1.
dist2 = 2.
plt.gca().add_patch(patches.Arc((x1-lc*sin(ang1), y1t-lc*cos(ang1)), 2*dist1, 2*dist1, theta1=90., theta2=-90.,lw=0.5,color='r'))
plt.gca().add_patch(patches.Arc((x2+lc*sin(ang1), y2t-lc*cos(ang1)), 2*dist1, 2*dist1, theta1=-90., theta2=90.,lw=0.5,color='r'))
plt.gca().add_patch(patches.Arc((x3-lc*sin(ang1), y3t-lc*cos(ang1)), 2*dist1, 2*dist1, theta1=90., theta2=-90.,lw=0.5,color='r'))

# teste de cota
plot_dimension(plt, (x2,y2t-lc), (x1,y1t-lc), offset=-1.5)
plot_dimension(plt, (x2,y2t-lc), (x3,y3t-lc), offset=1.5)
plot_dimension(plt, (x1,y1t-lc), (x3,y3t-lc), offset=-1.5)
plt.axis('equal')
plt.gca().set_ylim(17, ht+0.5)
plt.show()
```

```python
# Meio do vão
y1mv = y1 + h
y2mv = y2 + h
y3mv = y3 + h
l = f + lc
plt.figure()
plt.plot([x1, x1+l*sin(ang1)], [y1t, y1t-l*cos(ang1)],'ro-')
plt.plot([x2, x2+l*sin(ang1)], [y2t, y2t-l*cos(ang1)],'ro-')
plt.plot([x3, x3+l*sin(ang1)], [y3t, y3t-l*cos(ang1)],'ro-')
plt.gca().add_patch(patches.Arc((x1, y1t), 2*l, 2*l, theta1=-90., theta2=degrees(ang1)-90.,lw=0.5,color='r'))
plt.gca().add_patch(patches.Arc((x2, y2t), 2*l, 2*l, theta1=-90., theta2=degrees(ang1)-90.,lw=0.5,color='r'))
plt.gca().add_patch(patches.Arc((x3, y3t), 2*l, 2*l, theta1=-90., theta2=degrees(ang1)-90.,lw=0.5,color='r'))
plt.axis('equal')
plt.show()
```
