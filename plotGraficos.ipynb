## Fator de efetividade de vento no vão
O gráfico original (figura 17) vai até 800 m, mas usando a fórmula indiscriminadamente o gráfico volta...? Necessário saturar em um valor?

```python
L = np.arange(100,1200,5)
g = []
for i in L:
  temp = lambda i:nbr5422.fatVentoVao(i)
  g.append(temp(i))
plt.plot(L, g)
plt.grid()
plt.show()
```

```python
v = np.arange(7,22,0.5)
gc = -7.556e-5*v**3 + 7.867e-3*v**2 - 2.373e-1*v + 2.58
plt.plot(v, gc)
plt.grid()
plt.show()
```

```python
Lm = np.arange(200,800,5)
gl = 4e-10 * Lm**3 -5e-7 * Lm**2 -1e-4 * Lm + 1.0403
plt.plot(Lm, gl)
plt.grid()
plt.show()
```

```python
Vp = np.arange(0,30)
sigma = 2.25*(1-np.exp(-Vp**2/230))
plt.plot(Vp, sigma)
plt.grid()
plt.show()
```
